```python
import pandas as pd
import numpy as np
file_path = 'materiales (1).xlsx'
df = pd.read_excel(file_path)
print(df.head(16))

# Descripción general de los datos
print("\nDescripción del dataset:")
print(df.describe(include='all'))
pd.set_option('display.float_format', '{:.2f}'.format)
print("\nInformation del dataset:")
print(df.info())

```

```python
# Calcular medidas de tendencia central
# Media
mean_values = df.mean()

# Mediana
median_values = df.median()

# Moda (primera moda si hay varias)
mode_values = df.mode().iloc[0]

# Crear un DataFrame con los resultados
central_tendency = pd.DataFrame({
    "Mean": mean_values,
    "Median": median_values,
    "Mode": mode_values
})

# Mostrar los resultados
print("Medidas de Tendencia Central:")
print(central_tendency)
```

```python

import matplotlib.pyplot as plt

# Crear un boxplot para cada columna numérica
plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura según tus necesidades
df.boxplot()
plt.title('Boxplot de las Variables Numéricas')
plt.ylabel('Valores')
plt.xticks(rotation=45)  # Gira las etiquetas de las columnas si es necesario
plt.show()
```

```python
# Calcular los cuartiles y el rango intercuartílico para cada columna numérica
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Identificar los outliers
outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))

# Mostrar los outliers
print("Outliers detectados (True indica un outlier):")
print(outliers)

# Opcional: Mostrar las filas que contienen outliers
outliers_rows = df[outliers.any(axis=1)]
print("\nFilas con outliers:")
print(outliers_rows)
```

```python
# Varianza
variance = df.var()

# Desviación estándar
std_dev = df.std()

# Rango intercuartílico (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Rango (máximo - mínimo)
data_range = df.max() - df.min()

# Crear un DataFrame para mostrar los resultados
dispersion_measures = pd.DataFrame({
    "Varianza": variance,
    "Desviación Estándar": std_dev,
    "Rango Intercuartílico (IQR)": IQR,
    "Rango (Max - Min)": data_range
})

# Mostrar las medidas de dispersión
print("Medidas de Dispersión:")
print(dispersion_measures)
```
